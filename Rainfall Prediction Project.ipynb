```python
x.columns
```

```python
x_visualize = x.filter(['TempAvgF', 'DewPointAvgF', 'HumidityAvgPercent', 'SeaLevelPressureAvgInches', 'VisibilityAvgMiles', 'WindAvgMPH'])
```

```python
x_visualize
```

```python
cols_to_convert = ['DewPointAvgF', 'HumidityAvgPercent', 'SeaLevelPressureAvgInches', 'VisibilityAvgMiles',	'WindAvgMPH']
x_visualize[cols_to_convert] = x_visualize[cols_to_convert].astype(float)
```

```python
print("The Precipitation V/S the Attributes trend graph")
plt.figure(figsize=(10, 11))
for i in range(x_visualize.columns.size):
    plt.subplot(3, 2, i+1)
    
    plt.scatter(days, x_visualize[x_visualize.columns.values[i][:100]], color='g')
    plt.scatter(days[random_day], x_visualize[x_visualize.columns.values[i]][random_day], color='r')
    plt.title(x_visualize.columns.values[i])


plt.show()
    
```
