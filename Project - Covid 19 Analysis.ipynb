```python
Countries = list(Covid_Grouped.index)
```

```python
max_covid_rate = []
for each in Countries:
    max_covid_rate.append(Covid_Grouped.loc[each].diff().max())
    
Covid_Grouped['Maximum Rate of Covid Spread'] = max_covid_rate
```

```python
Covid_Grouped
```

```python
covid_data = pd.DataFrame(Covid_Grouped['Maximum Rate of Covid Spread'])
```

```python
covid_data
```
