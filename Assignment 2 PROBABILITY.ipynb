##### Two dices are rolled and computing the probability of getting the sum equals to 7 

```python
dice_1 = [1, 2, 3, 4, 5, 6]
dice_2 = [1, 2, 3, 4, 5, 6]
```

```python
# Calculate the possible events

length_possible_events = 0
for i in dice_1:
    for j in dice_2:
        if((i+j) == 7):
            length_possible_events = length_possible_events + 1
        else:
            continue
```

```python
length_possible_events
```

```python
#the sample space is 36 for 2 dice rolled
prob_sum_7 = (length_possible_events)/36
```

```python
prob_sum_7
```

### Conditional Probability and Bayes' Theorem

```python
# Num of favourable cases are elements ['Blue', 'Red']
# According to the conditional probability p(R|B) = (p(R) ∩ p(B)) / p(B))
# Sample space is 1000 records
```

```python
def simulate_draws(num_draws = 1000):
    balls = ['Red'] * 5 + ['Green'] * 7 + ['Blue'] * 8
    draws = []
    for _ in range(num_draws):
        draws.append(random.choice(balls))

    return draws
```

```python
draws = simulate_draws(1000)
```

```python
print(draws)
```

```python
def estimate_conditional_probability(draws):
    count_blue_then_red = 0
    count_prev_blue = 0

    for i in range(1, len(draws)):
        if draws[i-1] == 'Blue':
            count_prev_blue += 1
            if draws[i] == 'Red':
                count_blue_then_red += 1

    # Conditional Probability: P(Red | previous was Blue)
    
    return count_blue_then_red / count_prev_blue
```

```python
estimated_p_red_given_blue = estimate_conditional_probability(draws)
```

```python
print(f"The probability that the second ball drawn was red given that the first ball was blue is {estimated_p_red_given_blue}")
```

### Random Variables

```python
import numpy as np
```

```python
values = [1, 2, 3]
probabilities = [0.25, 0.35, 0.40]
sample_size = 1000
```

```python
sample = np.random.choice(values, size=sample_size, p=probabilities)
```

```python
empirical_mean = np.mean(sample)
empirical_variance = np.var(sample)
empirical_std_dev = np.std(sample)
```

```python
print(f"Empirical Mean: {empirical_mean}")
print(f"Empirical Variance: {empirical_variance}")
print(f"Empirical Standard Deviation: {empirical_std_dev}")
```

### Continuous Random Variables

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon
```

```python
mean = 5
sample_size = 2000
```

```python
samples = np.random.exponential(scale=mean, size=sample_size)
```

```python
x = np.linspace(0, np.max(samples), 1000)
pdf = expon.pdf(x, scale=mean)
```

```python
# Plot
plt.figure(figsize=(10, 6))
plt.hist(samples, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='black', label='Histogram')
plt.plot(x, pdf, 'r-', lw=2, label='Exponential PDF (mean=5)')
plt.title('Exponential Distribution: Histogram with PDF Overlay')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Central Limit Theorem

```python
# Generate 10,000 random numbers from a uniform distribution
population = np.random.uniform(low=0.0, high=1.0, size=10000)
```

```python
sample_size = 30
num_samples = 1000
sample_means = []
```

```python
for i in range(num_samples):
    sample = np.random.choice(population, size=sample_size, replace=True)
    sample_means.append(np.mean(sample))
```

```python
sample_means = np.array(sample_means)
```

```python
plt.figure(figsize=(12, 5))

# Plot the original uniform distribution
plt.subplot(1, 2, 1)
plt.hist(population, bins=30, color='lightgreen', edgecolor='black', density=True)
plt.title("Original Uniform Distribution")
plt.xlabel("Value")
plt.ylabel("Density")




# Plot the distribution of sample means
plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=30, color='skyblue', edgecolor='black', density=True)
plt.title("Distribution of Sample Means (n = 30)")
plt.xlabel("Sample Mean")
plt.ylabel("Density")

# Display the plots
plt.tight_layout()
plt.show()
```
